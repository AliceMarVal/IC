```python
import pandas as pd

file_path = 'Distâncias_aeroportos.csv'  # Caminho do arquivo
file_path1 = 'Nome_Aeroporto.csv'  # Caminho do arquivo
data = pd.read_csv(file_path)
data_nome = pd.read_csv(file_path1)
```

```python
df_a = pd.DataFrame(data_nome)
df = pd.DataFrame(data)
df_a.columns
```

```python
pd.set_option('display.max_rows', None)
```

```python
mapa_ids = {i+1: nome for i, nome in enumerate(df_a['Aeroportos de origem '])}
df['InputID'] = df['InputID'].replace(mapa_ids)
df['TargetID'] = df['TargetID'].replace(mapa_ids)
df['Emissão de CO2'] = df['Distância km'] * 0.123
df
```

```python
pd.reset_option('display.max_rows')
```

```python
import pandas as pd

# Lista de aeroportos indicados
aeroportos_interesse = [
    "Aeroporto Internacional de Guarulhos",
    "Aeroporto Internacional Guararapes - Gilberto Freyre",
    "Aeroporto Internacional Deputado Luís Eduardo Magalhães",
    "Aeroporto Internacional Governador Aluízio Alves",
    "Aeroporto Internacional Marechal Cunha Machado",
    "Aeroporto Internacional Pinto Martins"
]

# Resultado: Menores distâncias
resultados = []

# Iterar sobre todos os aeroportos de interesse
for aeroporto_interesse in aeroportos_interesse:
    # Filtrar distâncias válidas: aeroportos fora da lista para o aeroporto atual
    df_atual = df[
        (df['InputID'] == aeroporto_interesse) &
        (~df['TargetID'].isin(aeroportos_interesse))
    ]

    # Encontrar os 5 menores aeroportos para o aeroporto atual
    menores_distancias = df_atual.nsmallest(5, 'Distância km')

    # Adicionar os resultados ao DataFrame consolidado
    if not menores_distancias.empty:
        for _, linha in menores_distancias.iterrows():
            aeroporto_target = linha['TargetID']
            distancia = linha['Distância km']
            emissao = linha['Emissão de CO2']  # Emissões entre o aeroporto atual e o Target

            resultados.append({
                "Aeroporto Origem": aeroporto_interesse,  # Inclui o aeroporto de origem
                "Aeroporto Target": aeroporto_target,
                "Distância Origem-Target (Km)": distancia,
                "Emissão Total de CO2 (Kg)": emissao
            })

# Criar um DataFrame consolidado com os resultados
df_resultados = pd.DataFrame(resultados)

# Exibir o DataFrame com as colunas estruturadas
print(df_resultados)
```

```python
import pandas as pd

# Exemplo de dataset original
data = {
    'origem': ['Aeroporto1', 'Aeroporto1', 'Aeroporto2', 'Aeroporto2', 'Aeroporto3'],
    'destino': ['AeroportoA', 'AeroportoB', 'AeroportoA', 'AeroportoB', 'AeroportoA'],
    'distancia_km': [100, 200, 150, 250, 300],
    'emissao_co2': [10, 20, 15, 25, 30]
}
df = pd.DataFrame(data)

# Lista de aeroportos de destino de interesse
aeroportos_destino = ['AeroportoA', 'AeroportoB']

# Criar uma lista para armazenar as linhas do novo DataFrame
linhas = []

# Agrupar os dados por origem
for origem, group in df.groupby('origem'):
    row = {'origem': origem}
    for aeroporto in aeroportos_destino:
        # Filtrar os dados para o aeroporto de destino atual
        filtro = group[group['destino'] == aeroporto]
        if not filtro.empty:
            row[f'distancia_para_{aeroporto}'] = filtro['distancia_km'].values[0]
            row[f'emissao_para_{aeroporto}'] = filtro['emissao_co2'].values[0]
        else:
            row[f'distancia_para_{aeroporto}'] = None
            row[f'emissao_para_{aeroporto}'] = None
    linhas.append(row)

# Criar o novo DataFrame a partir das linhas
novo_df = pd.DataFrame(linhas)

novo_df
```
