```python
import pandas as pd

file_path = 'IMP.csv'  # Caminho do arquivo
data = pd.read_csv(file_path)
df = pd.DataFrame(data)
df = df.drop(columns=['Unnamed: 1'])
df = df.loc[df['Via'] == 'AEREA']
#Manter o Galeão e o Viracopos
```

```python
# Criar a matriz de associação
result = (
    df.assign(Valor=1)  # Adicionar uma coluna de valor como indicador
    .pivot_table(
        index='Bloco Econômico',  # Índice é o Bloco Econômico
        columns='URF',           # Colunas são os valores de URF
        values='Valor',          # Preenche com os valores da coluna "Valor"
        fill_value=0             # Substitui valores NaN por 0
    )
    .reset_index()               # Opcional: retorna o índice "Bloco Econômico" como coluna
)

result
```

```python
# Lista de URFs de interesse
urfs = [
    '0817600 - AEROPORTO INTERNACIONAL DE SAO PAULO/GUARULHOS',
    '0417901 - IRF - AEROPORTO INTERNACIONAL DOS GUARARAPES',
    '0517700 - AEROPORTO INTERNACIONAL DE SALVADOR DEP. LUIS EDUARDO',
    '0317700 - AEROPORTO INTERNACIONAL PINTO MARTINS',
    '0817700 - AEROPORTO INTERNACIONAL DE VIRACOPOS'
]

# Adicionar o cabeçalho "Bloco Econômico"
urfs = ['Bloco Econômico'] + urfs

# Filtrar as colunas de interesse
result = result.filter(items=urfs)

# Filtrar os blocos econômicos desejados
blocos_incluidos = ['Ásia (Exclusive Oriente Médio)', 'União Europeia - UE', 'Europa']
result = result.loc[result['Bloco Econômico'].isin(blocos_incluidos)].reset_index(drop=True)

result = pd.DataFrame(result)

# Exibir o DataFrame resultante
result.to_csv('Tabela_IMP.csv',index=False)
```
